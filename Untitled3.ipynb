functionalprogramming to print the n nantural numbers

````
def printNaturalNumbers(n):
  cnt=1;
  while(cnt<=n):
    print(cnt,end = " ");
    cnt=cnt+1;
  print()
  return
printNaturalNumbers(15);
````

````
def findfact(n):
  fact=1;
  while(n!=0):
    fact=fact*n;
    n=n-1;
  return fact;
print(findfact(6584));
````
